```python
# Distribution plot of a single column against different bankruptcy flag
def plotdist(column, h):
    sns.displot(df, x=column, hue=h, kind='kde')  
    plt.title(f'Distribution of {column} for Normal(0) vs Bankrupt(1) companies')
    plt.show()
```

```python
def plotbox(x, y):
    sns.boxplot(data=df, x=x, y=y, sym='')
    plt.title(f'IQR of {y} for Normal(0) vs Bankrupt(1) companies')
    plt.show()
```

```python
def calcMannWhitney(df_x, x, df_y, y):
    alpha = 0.05
    results = sp.stats.mannwhitneyu(x=df_x[x], y=df_y[y], \
                                alternative='two-sided', method='auto')
    print(f"U1 score: {results.statistic}, p-value: {results.pvalue}")
    if results.pvalue < alpha:
        print(f'We notice a significant difference of {x} between normal and bankrupt companies')
    else:
        print(f'There is no significant difference of {x} between normal and bankrupt companies')
```

```python
# Step 15: Create a crosstab between Flag and One If Total Liabilities Exceeds Total Assets Zero Otherwise
contingency_table = pd.crosstab(df['Flag'], df['One If Total Liabilities Exceeds Total Assets  Zero Otherwise'])
```

```python
# Step 16: Perform a chi square contingency test
# returns statistics, pvalue, degree of freedom (dof) and table of expected frequencies
print('Hypothesis:')
print('===========')
print('H0: Column \'Flag\' and column \'One If Total Liabilities Exceeds Total Assets Zero Otherwise\' independent from each other.')
print('H1: Otherwise not independent.\n')
stat, p, dof, expected = sp.stats.chi2_contingency(contingency_table)

print('dof=%d' % dof)
print(f'Expected frequencies:\n {expected}')

# interpret test-statistic
prob = 0.95
critical = sp.stats.chi2.ppf(prob, dof)
print('\nprobability=%.3f, critical=%.3f, stat=%.3f' % (prob, critical, stat))
if abs(stat) >= critical:
    print('Dependent (reject H0)')
else:
    print('Independent (fail to reject H0)')

# interpret p-value
alpha = 1.0 - prob
print('\nsignificance=%.3f, p=%.3f' % (alpha, p))
if p <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (fail to reject H0)')
```
